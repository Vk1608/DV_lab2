loading data

```python
url = 'https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/1_OneNum.csv'

df1 = pd.read_csv(url)
```

```python
df1.info()
```

```python
df1.max()
```

Histplot  with matplotlib limit for range

```python
fig, ax = plt.subplots()
sns.histplot(df1, ax=ax, binwidth=10,shrink=0.9)
ax.set_xlim(0, 300)
plt.show()
```

```python
fig, ax = plt.subplots()
sns.histplot(df1, ax=ax, binwidth=3,shrink=0.9)
ax.set_xlim(0, 300)
plt.show()
```

Density plot with pandas for range

```python
data=df1[df1['price']<300]
sns.kdeplot(data=data,x="price")
```

```python
data=df1[df1['price']<300]
sns.kdeplot(data=data,x="price", bw_adjust=.2)
```

two number

```python
url = 'https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/2_TwoNum.csv'

df2 = pd.read_csv(url)
```

```python
df2.shape
```

```python
df2.info()
```

Scatterplot

```python
plot_rate= df2.eval("SalePrice / GrLivArea").rename("plot_rate")
sns.scatterplot(data=df2,x="GrLivArea", y="SalePrice", hue=plot_rate)
```

```python

sns.regplot(data=df2,x="GrLivArea", y="SalePrice",color="teal",line_kws={"color": "black"})
```

```python
sns.lmplot(data=df2,x="GrLivArea", y="SalePrice",order=3)
```

Overplotting

```python
data=df2[df2['GrLivArea']<3000]
data=data[data['SalePrice']<350000]
sns.kdeplot(x=data.GrLivArea, y=data.SalePrice)
plt.show()
```

```python
# Custom the color, add shade and bandwidth
sns.kdeplot(x=data.GrLivArea, y=data.SalePrice, cmap="Reds", shade=True, bw_adjust=.5)
plt.show()

# Add thresh parameter
sns.kdeplot(x=data.GrLivArea, y=data.SalePrice, cmap="Blues", shade=True, thresh=0)
plt.show()
```

```python
sns.displot(x=df2.GrLivArea, y=df2.SalePrice)
```
